Assign a name for each group rather than dealing with a number ðŸ¤·

```python
age_rename_map = {
    '10-19': 'Teens',
    '20-29': 'Young Adults',
    '30-39': 'Adults',
    '40-49': 'Midlife Adults',
    '50-59': 'Older Adults'
}

data['Age Group Label'] = data['Age'].map(age_rename_map)
```

```python
from scipy.stats import chi2_contingency

ct_usage = pd.crosstab(data['Age Group Label'], data['How often do you use chatbots?'])

age_order = ['Teens', 'Young Adults', 'Adults', 'Midlife Adults', 'Older Adults']
usage_order = ['Daily', 'Several times a week', 'Several times a month', 'Rarely']
ct_usage = ct_usage.reindex(index=age_order, columns=usage_order, fill_value=0)

chi2, p, dof, expected = chi2_contingency(ct_usage)

print("Chi-Square Test Ergebnisse:")
print(f"ChiÂ²-Wert: {chi2:.2f}")
print(f"p-Wert: {p:.4f}")
print(f"Freiheitsgrade: {dof}")
```

## Add new columns for each of tasks

```python
data['For which tasks do you use chatbots? '].value_counts()
```

Append these task as column to the main dataframe

```python
def assign_tasks(task_string: str) -> list[bool]:
    task_string = str(task_string).strip()

    # Split into individual tasks (e.g., 'Programming;Writing' â†’ ['Programming', 'Writing'])
    tasks = [t.strip() for t in task_string.split(";") if t.strip()]

    # Initialize dictionary of flags
    known_tasks = {
        "Explanation",
        "Teaching",
        "Programming",
        "Creative tasks",
        "Writing",
    }
    result = {
        "Explanation": False,
        "Teaching": False,
        "Programming": False,
        "Creative tasks": False,
        "Writing": False,
        "Other": False,
    }

    for task in tasks:
        if task in known_tasks:
            result[task] = True
        else:
            result["Other"] = True

    return list(result.values())
```

```python
tasks_col = 'For which tasks do you use chatbots? '

task_columns = ['Explanation', 'Teaching', 'Programming',\
                'Creative tasks', 'Writing', 'Other']

task_flags = data[tasks_col].apply(assign_tasks)

task_flags_df = pd.DataFrame(task_flags.tolist(), columns=task_columns, index=data.index)

data = pd.concat([data, task_flags_df], axis=1)
```

## Wie BildungsabschlÃ¼sse die Nutzung von Chatbots beeinflussen

```python
data.groupby('Education (highest degree)')[task_columns].sum()
```

* Let's focus on the education levels that have significant participation.

```python
# Filter only High school diploma or equivalent, Bachelor's and Master's degree
edu_formal_degrees = data[
    data['Education (highest degree)'].isin([
        "High school diploma or equivalent",
        "Bachelor's degree",
        "Master's degree"
    ])
]
```

```python
# Group task usage
task_usage = edu_formal_degrees.groupby('Education (highest degree)')[task_columns].sum()

# Define desired order
ordered_levels = [
    "High school diploma or equivalent",
    "Bachelor's degree",
    "Master's degree"
]

# Reindex to match desired order
task_usage = task_usage.reindex(ordered_levels)
```

```python
task_usage
```

```python
task_usage.T.plot(kind='bar', figsize=(10, 6), color=plt.get_cmap('Dark2').colors)

plt.title("Chatbot Task Usage by Education Level")
plt.ylabel("Number of Users")
plt.xlabel("Task")
plt.xticks(rotation=45)
plt.legend(title="Education Level")
plt.tight_layout()
plt.show()
```

## HÃ¤ufigkeit der Chatbot-Nutzung fÃ¼r Programmierung und kreative Aufgaben in verschiedenen Altersgruppe

```python
data.groupby('Age')[['Programming', 'Creative tasks']].sum()
```

```python
import matplotlib.pyplot as plt

# Define your mapping
age_rename_map = {
    '10-19': 'Teens',
    '20-29': 'Young Adults',
    '30-39': 'Adults',
    '40-49': 'Midlife Adults',
    '50-59': 'Older Adults'
}

# Group and sum, then rename the index
age_task_counts = data.groupby('Age')[['Programming', 'Creative tasks']].sum()
age_task_counts = age_task_counts.rename(index=age_rename_map)

# Plot
age_task_counts.plot(kind='bar', figsize=(10, 6), color=['#1f78b4', '#984ea3'])

plt.title("Chatbot Usage by Age Group: Programming vs. Creative Tasks")
plt.xlabel("Age Group")
plt.ylabel("Number of Users")
plt.xticks(rotation=45)
plt.legend(title="Task")
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.heatmap(age_task_counts, annot=True, cmap='Blues', fmt='d')
plt.title("Heatmap: Chatbot Task Usage by Age Group")
plt.ylabel("Age Group")
plt.xlabel("Task")
plt.tight_layout()
plt.show()
```

```python
# Prozentuale Chatbot-Nutzung nach Altersgruppe: Programmierung und kreative Aufgaben
percent_df = age_task_counts.div(age_task_counts.sum(axis=1), axis=0) * 100
percent_df = percent_df.fillna(0).round(1)

print(percent_df)
```

```python
percent_df.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='cividis')
plt.ylabel('Percentage')
plt.title('Verteilung der Chatbot-Aufgaben nach Altersgruppen (%)')
plt.legend(title='Aufgabe', bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()
```
